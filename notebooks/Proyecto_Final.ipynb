## Preparamos los datos para el modelo, ya que consideramos que estan limpios

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
```

#### Separación predictoras y target

```python
eph2022.dtypes
```

```python
dummie_datos = pd.get_dummies(eph2022.drop(['Ocupados_no_agropecuarios','Años_de_estudio', 'Edad', 'Años_está_sin_trabajo', 'Ing_por_actividad_principal_corregido'], axis=1))
numericos = eph2022[['Ocupados_no_agropecuarios','Años_de_estudio', 'Edad', 'Años_está_sin_trabajo', 'Ing_por_actividad_principal_corregido']]
df_final = pd.concat([dummie_datos, numericos], axis=1)
```

```python
df_final.head()
```

```python
X = df_final.drop('Ocupados_no_agropecuarios', axis = 1)
y = df_final['Ocupados_no_agropecuarios']
```

Balanceamos los datos

```python
smt = SMOTE(random_state=123)
X, y = smt.fit_resample(X, y)
```

```python
df_final = pd.concat([X, y], axis=1)
```

```python
ax = sns.countplot(x='Ocupados_no_agropecuarios', data=df_final)
```

#### Instanciamos las matrices de entrenamiento y prueba

```python
X = df_final.drop('Ocupados_no_agropecuarios', axis = 1)
y = df_final['Ocupados_no_agropecuarios']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
norm = StandardScaler()
X_normalizado = norm.fit_transform(X)
X_normalizado
```

```python
X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(X_normalizado, y, test_size=0.3, random_state=123)
```

#### Creamos el modelo y lo entrenamos

### Comenzamos con KNeighbors

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(metric='euclidean')
```

```python
knn.fit(X_entrenamiento, y_entrenamiento)
```

```python
predict_knn = knn.predict(X_prueba)
```

```python
train_acc_log = round((knn.score(X_entrenamiento, y_entrenamiento))*100,1)
test_acc_log = round((knn.score(X_prueba, y_prueba))*100,1)

print("Training accuracy: %s %%" % train_acc_log)
print('Test accuracy: %s %%' % test_acc_log)
print('\nMatriz de confusion: ')
print(confusion_matrix(y_prueba, predict_knn))
print('\nReporte de Clasificacion :')
print(classification_report(y_prueba, predict_knn))
```

### Regresion Logistica

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(penalty="l2",C=.1,random_state=0)
lr.fit(X_entrenamiento,y_entrenamiento)
predict_lr = lr.predict(X_prueba)

train_acc_log = round((lr.score(X_entrenamiento, y_entrenamiento))*100,1)
test_acc_log = round((lr.score(X_prueba, y_prueba))*100,1)

print("Training accuracy: %s %%" % train_acc_log)
print('Test accuracy: %s %%' % test_acc_log)
print('\nMatriz de confusion: ')
print(confusion_matrix(y_prueba, predict_lr))
print('\nReporte de Clasificacion :')
print(classification_report(y_prueba, predict_lr))
```
